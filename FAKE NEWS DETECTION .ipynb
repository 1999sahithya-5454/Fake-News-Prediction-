## **Importing Necessary Libraries**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
```

### **IMPORTING THE DATASET**


```python
dataset = pd.read_csv('/content/news.csv', sep = ',', encoding = 'iso-8859-1', squeeze =" false", engine="python")
dataset.head(6)
```

## **Splitting the dataset into X & y**

```python
#Splitting for simplicity purpose
#text/news & label for it whether its fake or real
X = dataset['text']
y = dataset['label']
```

```python
X
```

```python
y
```

## **Splitting the dataset into training & testing sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## **Vectorization**

```python
tfvect = TfidfVectorizer(stop_words = 'english', max_df = 0.7)
tfid_X_train = tfvect.fit_transform(X_train)
tfid_X_test = tfvect.transform(X_test)
```

## **Machine Learning Model**

```python
#Classifying fake & real news
classifier = PassiveAggressiveClassifier(max_iter = 50)
classifier.fit(tfid_X_train, y_train)
```

### **Model Accuracy**

```python
y_pred = classifier.predict(tfid_X_test)
score = accuracy_score(y_test, y_pred)

print(f'Accuracy: {round(score * 100, 2)}%')
```

### **Making a confusion matrix for classification visibilty**

```python
cf = confusion_matrix(y_test, y_pred, labels = ['FAKE', 'REAL'])
print(cf)
```

### **Creating a function for testing the model on dat**a

```python
def fake_news_detection(news):
  input_data = [news]
  vectorized_input_data = tfvect.transform(input_data)
  prediction = classifier.predict(vectorized_input_data)
  print(prediction)
```

# **Results**

```python
fake_news_detection('10 Ways America Is Preparing for World War 3') #randomly taken
```
